```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x=[3,4,5,6,7]
y=[8,2,3,7,5]
plt.plot(x,y,label="f(x)",color="g",linestyle="--",marker="*")
plt.xticks([1,2,3,4,5,6,7])
plt.yticks([2,3,4,5,6,7,8])
#label
plt.xlabel("x-axis")
plt.ylabel("y-axis")
#title
plt.title("samle graph")
# legend
plt.legend()
#resize graph
plt.figure(figsize=(3,5),dpi=500)
#save graph
plt.savefig('sampleg.png',dpi=500)
plt.show()
```
