# Using Pandas Dataframe

```python
df=pd.read_csv("car-sales.csv")#importing an csv/excel file
df
```

```python
test = pd.Series(np.random.randn(1000),
                 index=pd.date_range("1/1/2020",periods=1000))

#random numbers will be generated with respect to the date

test=test.cumsum() #cumulative sum
test.plot()
```

```python
df
```

```python
#changing price column to interger from char
df["Price"] = df["Price"].str.replace('[\$\,\.]','')# 2 EXTRA ZEROS DUE TO DECIMAL POINT REMOVAL
df["Price"] = df["Price"].str[:-2]
df #if it removes more than 2 place try rerunning the import csv part
```

```python
df["Sale_date"] = pd.date_range("1/1/2020",periods=len(df))
df #adding a new column 
```

```python
#ADDING NEW COLUMN FOR TOTAL SALES AT THE EN OD RESPECTIVE DATE
#WE HAVE TO FIND THE CUMULATIVE SUM
#Price are still str so convert to int

df["Total"] = df["Price"].astype(int).cumsum()
df
```

```python
df.dtypes #check the datatypes
```

```python
df.plot(x="Sale_date",y="Total");
```

```python
df["Price"]= df["Price"].astype(int)
df.plot(x="Odometer (KM)",y="Price",kind="scatter");
```

```python
heart_d = pd.read_csv("heart-disease.csv")
heart_d
```

```python
#ploting all the data in single line
#athe scale of the graph will be the same so it will be clumsy
```

```python
heart_d.plot.hist(figsize=(10,30),subplots="True");
```

```python
# seperate the patients with age more than 50
over_50 = heart_d[heart_d["age"] > 50]
over_50
```

```python
#pyplot method
over_50.plot(kind='scatter',x='age',y='chol',c='target');
# c is the colour based on "target" in the dataframe
```

```python
fig,ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
#ax.set_xlim([45,90])
```

```python
fig , ax = plt.subplots(figsize=(10,6))
scat = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

ax.set(title="age vs cholestrol level",xlabel="AGE",ylabel="CHOLESTROL");
ax.legend(*scat.legend_elements(),title= "target");
```

```python
# a line to indicate the avg 
```

```python
fig , ax = plt.subplots(figsize=(10,6))
scat = ax.scatter(x=over_50["age"],y=over_50["chol"],c=over_50["target"]);

ax.set(title="age vs cholestrol level",xlabel="AGE",ylabel="CHOLESTROL");
ax.legend(*scat.legend_elements(),title= "target");

ax.axhline(over_50["chol"].mean(),linestyle="--");
```

```python
#chol,age,thalach

fig , (ax1,ax2) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50['target'].map({1:'black',0:'deeppink'}));

ax1.set(title="AGE VS CHOLESTROL",xlabel="AGE",ylabel="CHOLESTROL")
ax1.axhline(y=over_50["chol"].mean(),linestyle="--")

scatter1 = ax2.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50['target'].map({1:'black',0:'greenyellow'}
                                             ));

ax2.set(title="AGE VS heartbeat",xlabel="AGE",ylabel="heartbeat")
ax2.axhline(y=over_50["thalach"].mean(),linestyle=":")


#title for over all fig

fig.suptitle("HEART DISEASE ANALYSIS",fontsize=16,fontweight="bold");
```

## CUstomizing the grid

```python
#chol,age,thalach

fig , (ax1,ax2) = plt.subplots(nrows=2,ncols=1,figsize=(10,10))

scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50['target'],cmap="magma");
ax1.set(title="AGE VS CHOLESTROL",xlabel="AGE",ylabel="CHOLESTROL")
ax1.axhline(y=over_50["chol"].mean(),linestyle="--")
ax1.legend(*scatter.legend_elements(),title= "target");
ax1.set_xlim([50,80])
ax1.set_ylim([10,600])



scatter1 = ax2.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50['target'],cmap="winter");
ax2.set(title="AGE VS heartbeat",xlabel="AGE",ylabel="heartbeat")
ax2.axhline(y=over_50["thalach"].mean(),linestyle=":")
ax2.legend(*scatter1.legend_elements(),title= "target");
ax2.set_xlim([50,80])
ax2.set_ylim([60,200])
#title for over all fig

fig.suptitle("HEART DISEASE ANALYSIS",fontsize=16,fontweight="bold");
```
